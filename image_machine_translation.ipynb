```python
import cv2
import pytesseract
from PIL import Image
pytesseract.pytesseract.tesseract_cmd = r'E:/Tesseract-OCR/tesseract.exe'
from googletrans import Translator
translator = Translator()

from mtranslate import translate
```

```python
window_name = "Frame"
#cv2.namedWindow(window_name, cv2.WND_PROP_FULLSCREEN)
#cv2.setWindowProperty(window_name, cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)
                      
# define a video capture object
vid = cv2.VideoCapture(0)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, 7200)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, 7200)
pts=[]
while(True):
      
    # Capture the video frame
    # by frame
    ret, frame = vid.read()
    
    #print(frame.shape)
    text=pytesseract.image_to_string(cv2.cvtColor(frame[400:,400:1200,:], cv2.COLOR_BGR2RGB), lang='tam')
    if (len(text)>10):
        pt=translate(text,"en",'ta')
        pt=str(pt).strip().encode("ascii","ignore")
        pt=str(pt)
        pts.append(pt)
    else:
        pt="..."
    
    # Display the resulting frame
    cv2.putText(frame,pt, (25,150),cv2.FONT_HERSHEY_SIMPLEX,2,(0, 41, 256,100), 4)
    #cv2.rectangle(frame, (800,200), (1500,600), -1)
    cv2.imshow(window_name, frame)
    cv2.imwrite("frame.jpg",frame)
    
    # the 'q' button is set as the
    # quitting button you may use any
    # desired button of your choice
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
  
# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
